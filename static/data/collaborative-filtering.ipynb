```python
import numpy as np

ratings = np.zeros((840, 556)) # (|profiles|, |drinks|) when ratings were recorded
columns = {}
index = 0
with open('ratings.db') as f:
    for line in f.readlines():
        terms = line.split(',')
        if terms[1] not in columns:
            columns[terms[1]] = index
            index += 1
        ratings[int(terms[0]), columns[terms[1]]] = int(terms[2])
```

```python
masked = np.ma.masked_array(ratings, mask=(ratings == 0), fill_value=0)
average = np.average(masked, axis=1)
ratings = np.array(masked - average.reshape((average.shape[0], 1)))
```

```python
profile_ratings = np.zeros(556)
profile_ratings[0] = 1
profile_ratings[2] = -1
profile_ratings[4] = 1
profile_masked = np.ma.masked_array(profile_ratings, mask=(profile_ratings == 0), fill_value=0)
profile_average = np.average(profile_masked)
profile_ratings = np.array(profile_masked - profile_average)
profile_ratings
```
